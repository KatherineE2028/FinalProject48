```python
import pandas as pd
speed = pd.read_csv("Speed.tsv", sep="\t")
speed.head()
```

## Introduction ##
For my program, I decided to rank the neighborhoods based on their speed limit keeping in mind that a slower speed limit would lead to fewer car accidents and thus a safer neighborhood. I started by importing my data set and using the ".groupby( )" function to organize it by neighborhood and average speed limit and then organizing them from least to worst and put the areas without a speed limit put at the bottom, thinking that they would be more likely to cause accidents. 

```python
speed['speed_limit'] = pd.to_numeric(speed['speed_limit'], errors='coerce')
mean_speed = speed.groupby('neighborhood')['speed_limit'].mean()
sorted_mean_speed = mean_speed.sort_values()

pd.set_option('display.max_rows', None)
sorted_mean_speed = sorted_mean_speed.sort_values().reset_index()
sorted_mean_speed.index = sorted_mean_speed.index + 1
sorted_mean_speed.index.name = '#'
print((sorted_mean_speed))
```


Ultimately the optimum neighborhoods ended up being Homewood South and Hazlewood with an average speed limit just above 21mph, I think this method of ranking makes the most sense as it puts the neighborhoods in a quantifiable order that removes any subjectivity or interperatation while following a proper reasoning that leads to a direct answer.
