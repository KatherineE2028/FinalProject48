```python
import pandas as pd
speed = pd.read_csv("Speed.tsv", sep="\t")
speed.head()
```

## Introduction ##
For my program, I decided to rank the neighborhoods based on their speed limit keeping in mind that a slower speed limit would lead to fewer car accidents and thus a safer neighborhood. I started by importing my data set and using the ".groupby( )" function to organize it by neighborhood and ".mean( )" for the average speed limit and then organizing them from least to worst while putting the areas without a speed limit put at the bottom with the assumption that they would be more likely to cause accidents. 

```python
speed['speed_limit'] = pd.to_numeric(speed['speed_limit'], errors='coerce')

#Organzie dataset by neighborhood and display its average speed limit
mean_speed = speed.groupby('neighborhood')['speed_limit'].mean()

#Sort the neighborhoods from those with the lowest speed limits to those with the highest
sorted_mean_speed = mean_speed.sort_values()

#Create a rank collumn to give each neighborhood a quantifiable rank value
sorted_mean_speed = sorted_mean_speed.sort_values().reset_index()
sorted_mean_speed.index = sorted_mean_speed.index + 1
sorted_mean_speed.index.name = '#'

#Displays ordered mea
print((sorted_mean_speed))
```

## Conclusion ##
Ultimately the optimum neighborhoods based upon their speed limits ended up being Homewood South and Hazlewood with an average speed limit just above 21mph, I think this method of ranking makes the most sense as it puts the neighborhoods in a quantifiable order that removes any subjectivity or interperatation while following a proper reasoning that leads to a direct answer.
